## Convert imbalanced dataset into balanced dataset using Up/Over sampling technique - RandomOverSampler

Data Source : https://www.kaggle.com/mlg-ulb/creditcardfraud

Similar to under or down sampling techniques , Over sampling is a way to balance the output class distribution of an imbalanced dataset. One of the upsampling technique is to perform random oversampling which supplements training data with multiple copies of some of the minority classes instead of repeating every sample in the minority class.

There are other different oversampling techniques like SMOTE (Synthetic Minority Over-sampling Technique) ,ADASYN (Adaptive synthetic),  Class-weight using extrapolation etc

Here we will explore Random Oversampling technique using imblearn package

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:\\EBS\\0. Python\\Data Science with Python\\Handle imbalanced dataset\\creditcard.csv",sep=",")
```

```python
# See the sample dataset
df.head()
```

```python
# Information about dataset
df.info()
```

```python
# Basic Statistical information about independent and dependent features
df.describe()
```

```python
# Check if there is any null value in any feature
df.isnull().values.any()
```

```python
# Split independent and dependent features in 2 different DataFrame
X = df.drop('Class',axis=1)
y = df['Class']
```

```python
# Independent Features
X
```

```python
# Dependent or Output Feature
y
```

```python
# Check unique count of each class(0 = Genuine, 1 = Fraud) in output label
df['Class'].value_counts()
```

```python
# Plotting graph to distribute Class population
df['Class'].value_counts().plot(kind='bar')
plt.title("Genuine and Fraudulent Class Distribution")
plt.xticks([0,1],['Genuine','Fraudulent'])
plt.xlabel('Class')
plt.ylabel('Count')
```

```python
# Apply over sampling technique
from imblearn.over_sampling import RandomOverSampler
os = RandomOverSampler(sampling_strategy=1.0) # 1.0 makes Minority class samples count same as Mjority class count
# os = RandomOverSampler(sampling_strategy=.5) # .5 makes the Minority class samples count 50% as Mjority class count
X_resample, y_resample = os.fit_sample(X,y)
```

```python
# Check the resample count 
y_resample.value_counts()
```

```python
type(X_resample)
```

```python
# Check the Fraud and Genuine data distribution after over sampling
from collections import Counter
print('Actual dataset {}'.format(Counter(y)))
print('Undersampled dataset {}'.format(Counter(y_resample)))
```

```python
# Plotting 2-Class distribution after over sampling 
y_resample.value_counts().plot(kind='bar')
plt.title("Genuine and Fruadulent Class Distribution")
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks([0,1],['Genuine','Fraudulent'])
```
